```python
import pandas as pd
import requests
```

```python
responses = [requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}').json()['results'] for i in range(1,10)]
```

```python
df = pd.concat(list(map(lambda x: pd.DataFrame(x)[['id','title','overview','release_date','popularity','vote_average','vote_count']], responses)))
```
